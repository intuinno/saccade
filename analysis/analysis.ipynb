```python
import pickle
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('sac-010-rev2-5M.pickle', 'rb') as f:
    random_exploration = pickle.load(f)

with open('sac-010-rev2-p2e-5M.pickle', 'rb') as f:
    p2e_da = pickle.load(f)

with open('sac-011-rev0-5M.pickle', 'rb') as f:
    p2e_pe = pickle.load(f)

with open('sac-012-rev0-2M.pickle', 'rb') as f:
    ac_pe = pickle.load(f)
```

```python
ac_pe.keys()
```

```python
t = 2652500
```

```python
ac_pe_t = ac_pe[t].numpy()
random_exploration_t = random_exploration[2652516].numpy()
p2e_da_t = p2e_da[2652516].numpy()
p2e_pe_t = p2e_pe[t].numpy()
```

```python
plt.figure(figsize=(4, 3))
axes = plt.gca()
plt.plot(random_exploration_t, label="random")
plt.plot(ac_pe_t, label="AC-PE")
# plt.plot(p2e_da_t, label="P2E-DA")
plt.plot(p2e_pe_t, label="P2E-PE")

plt.legend()
plt.xlabel("Time Steps")
plt.ylabel("MSE (Lower is better)")
plt.title("Scene Convergence Time")

fig=plt.gcf()
plt.show()
fig.savefig(f'sct_comparison.pdf', bbox_inches='tight')
```

```python

import csv

def csv_to_ndarray(file_path):
    # Use numpy's genfromtxt to read the CSV file
    data = np.genfromtxt(file_path, delimiter=',', skip_header=1)
    return data
```

```python
wct_ac = csv_to_ndarray('wct-ac.csv')
wct_p2e = csv_to_ndarray('wct-p2e.csv')
wct_random = csv_to_ndarray('wct-random.csv')
```

```python
wct_ac.shape, wct_p2e.shape, wct_random.shape
```

```python
plt.figure(figsize=(4, 3))
axes = plt.gca()
plt.plot(wct_random[:55,1], wct_random[:55,2], label="random")
plt.plot(wct_ac[:55,1],wct_ac[:55,2], label="AC-PE")
plt.plot(wct_p2e[:55,1], wct_p2e[:55,2], label="P2E-PE")

# plt.plot(p2e_pe_t, label="P2E-PE")

plt.legend()
plt.xlabel("Training Steps")
plt.ylabel("MSE (Lower is better)")
plt.title("World model Convergence Time")

fig=plt.gcf()
plt.show()
fig.savefig(f'wct_comparison.pdf', bbox_inches='tight')
```
